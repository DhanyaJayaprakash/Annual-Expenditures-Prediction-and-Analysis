## Feature Selection Using Correlation Method

```python
Exp_Features=expenditure_sample.drop('Target',axis=1)
Exp_Target=expenditure_sample['Target']
```

### Normalising Data

```python
from sklearn.preprocessing import StandardScaler

scale= StandardScaler()
std_Features= scale.fit_transform(Exp_Features) 
print(std_Features)
```

```python
plt.figure(figsize=(12,10))
cor = Exp_Features.corr()
```

```python
top_corr_features = cor.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(expenditure_sample[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
def correlation(dataset, threshold):
    colcorr = set()  # Set of all the names of correlated columns
    corrmatrix = expenditure_sample.corr()
    for i in range(len(corrmatrix.columns)):
        for j in range(i):
            if abs(corrmatrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corrmatrix.columns[i]  # getting the name of column
                colcorr.add(colname)
    return colcorr
```

```python
corrfeatures = correlation(Exp_Features, 0.1)
len(set(corrfeatures))
```

```python
corrfeatures
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(std_Features,Exp_Target,test_size=.2)
```

# Model Creation

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_test, y_test)
```

```python
reg.score(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
Linear_Prediction=reg.predict(X_train)
```

```python
Linear_Prediction
```

#### Linear Regression Evaluation

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, Linear_Prediction) #mean square error calculation
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, Linear_Prediction) #r square error calculation
```
